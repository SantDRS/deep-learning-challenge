## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf
```

```python
#  Import and read the charity_data.csv.
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_df.drop(['EIN', 'NAME'], axis=1, inplace=True)
application_df.head()
```

```python
# Determine the number of unique values in each column.
unique_counts = application_df.nunique()
print(unique_counts)
```

```python
# Look at APPLICATION_TYPE value counts for binning
application_types_count = application_df['APPLICATION_TYPE'].value_counts()
print(application_types_count)
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`

cutoff_value = 200

application_types_to_replace = list(application_types_count[application_types_count < cutoff_value].index)

# Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
classification_types_count = application_df['CLASSIFICATION'].value_counts()
print(classification_types_count)
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
classification_counts_filtered = classification_types_count[classification_types_count > 1]
print(classification_counts_filtered)
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
classifications_cutoff_value = 1500

classifications_to_replace = list(classification_types_count[classification_types_count < classifications_cutoff_value].index)

# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")
    
# Check to make sure binning was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`

# Select the categorical columns to be converted

# Perform one-hot encoding using pd.get_dummies
application_dummies = pd.get_dummies(application_df)

# Concatenate the on-hot encoded columns with the original DF
#application_df = pd.concat([application_df, one_hot_encoded], axis=1)

# Drop the original categorical columns
#application_df.drop(categorical_columns, axis=1, inplace=True)

```

```python
# Display the updated DF
application_dummies.head()
```

```python
# Split our preprocessed data into our features and target arrays
# Split the DF into features (X) and target (y)
X = application_dummies.drop('IS_SUCCESSFUL', axis=1).values # features all columns except target column
y = application_dummies['IS_SUCCESSFUL'].values # target column

# Display the shapes of the features and target arrays
print('Shape of X:', X.shape)
print('Shape of y:', y.shape)
```

```python
# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

# Display the shapes of the training and testing datasets
print('Shape of X_train:', X_train.shape)
print('Shape of X_test:', X_test.shape)
print('Shape of y_train:', y_train.shape)
print('Shape of y_test:', y_test.shape)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
X_train_scaled.shape
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
number_input_features = X_train_scaled.shape[1]
hidden_nodes_layer1 =  90
hidden_nodes_layer2 = 30

nn_model_1 = tf.keras.models.Sequential()

# First hidden layer
nn_model_1.add(tf.keras.layers.Dense(units=hidden_nodes_layer1, input_dim=number_input_features, activation="relu"))

# Second hidden layer
nn_model_1.add(tf.keras.layers.Dense(units=hidden_nodes_layer2, activation="relu"))

# Output layer
nn_model_1.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn_model_1.summary()
```

```python
# Compile the model
nn_model_1.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model
fit_model = nn_model_1.fit(X_train_scaled,y_train,epochs=100)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn_model_1.evaluate(X_test_scaled, y_test, verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Second attempt adding a third layer with more nodes
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
number_input_features = X_train_scaled.shape[1]
hidden_nodes_layer1 =  90
hidden_nodes_layer2 = 60
hidden_nodes_layer3 = 30

nn_model_2 = tf.keras.models.Sequential()

# First hidden layer
nn_model_2.add(tf.keras.layers.Dense(units=hidden_nodes_layer1, input_dim=number_input_features, activation="relu"))

# Second hidden layer
nn_model_2.add(tf.keras.layers.Dense(units=hidden_nodes_layer2, activation="relu"))

# Third hidden layer
nn_model_2.add(tf.keras.layers.Dense(units=hidden_nodes_layer3, activation="relu"))

# Output layer
nn_model_2.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn_model_2.summary()
```
